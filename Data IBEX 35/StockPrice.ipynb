```python
import pandas as pd
import yfinance as yf

df = pd.read_csv('Ticketsvalidos.csv')
index_df = pd.read_csv('stockindex.csv')

start_date = '2020-01-01'
end_date = '2024-12-31'

for _, row in index_df.iterrows():
    sufijo = row['Sufijo']
    pais = row['País']

    tickers = df[df['ticker_market'].str.endswith(sufijo)]['ticker_market'].tolist()
    if not tickers:
        continue

    data = yf.download(tickers, start=start_date, end=end_date, group_by='ticker', auto_adjust=False)
    if data.empty:
        continue

    data = data.stack(level=0).rename_axis(['FECHA', 'ticker']).reset_index()
    stockprice = data[["ticker", "Close", "High", "Low", "Open", "Volume", "FECHA"]]
    stockprice['FECHA'] = pd.to_datetime(stockprice['FECHA']).dt.tz_localize(None)

    filename = f"stockprice_{pais.replace(' ', '_')}.xlsx"
    stockprice.to_excel(filename, index=False)
    print(f"{pais} → {filename}")
```

```python
import zipfile
import os

# Lista de archivos Excel generados
excel_files = [f for f in os.listdir() if f.startswith('stockprice_') and f.endswith('.xlsx')]

with zipfile.ZipFile("stockprices_all.zip", "w") as zipf:
    for file in excel_files:
        zipf.write(file)
        os.remove(file)  # Elimina el archivo Excel si ya está en el zip (opcional)

print("✅ Todos los archivos Excel han sido comprimidos en 'stockprices_all.zip'")
```

```python
#Francia
import pandas as pd
import yfinance as yf
from datetime import datetime

df = pd.read_csv('Ticketsvalidos.csv')
start_date = '2020-01-01'
end_date = '2024-12-31'


tickers_pa = df[df['ticker_market'].str.endswith('.PA')]['ticker_market'].tolist()

data = yf.download(tickers_pa, start=start_date, end=end_date)
df_long = data.stack(level=1).reset_index()
df_long.rename(columns={'level_1': 'ticker_market', 'Date': 'FECHA'}, inplace=True)

stockprice_pa = df_long[["Ticker", "Close", "High", "Low", "Open", "Volume", "FECHA"]]

ibex = yf.Ticker('^FCHI')
ibex_hist = ibex.history(start=start_date, end=end_date)

if ibex_hist.empty:
    ibex_hist = ibex.history(period='5d').tail(1)

ibex_open = ibex_hist['Open'].values[0]
ibex_close = ibex_hist['Close'].values[0]

stockprice_pa['StockIndexOpen'] = ibex_open
stockprice_pa['StockIndexClose'] = ibex_close

stockprice_pa.to_excel('stockprice_pa.xlsx', index=False)

stockprice_pa.head()
```
