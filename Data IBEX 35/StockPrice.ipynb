```python
#espaÃ±a
import pandas as pd
import yfinance as yf
from datetime import datetime

df = pd.read_csv('Ticketsvalidos.csv')
start_date = '2020-01-01'
end_date = '2024-12-31'


tickers_mc = df[df['ticker_market'].str.endswith('.MC')]['ticker_market'].tolist()

data = yf.download(tickers_mc, start=start_date, end=end_date)
df_long = data.stack(level=1).reset_index()
df_long.rename(columns={'level_1': 'ticker_market', 'Date': 'FECHA'}, inplace=True)

stockprice_mc = df_long[["Ticker", "Close", "High", "Low", "Open", "Volume", "FECHA"]]

ibex = yf.Ticker('^IBEX')
ibex_hist = ibex.history(start=start_date, end=end_date)

if ibex_hist.empty:
    ibex_hist = ibex.history(period='5d').tail(1)

ibex_open = ibex_hist['Open'].values[0]
ibex_close = ibex_hist['Close'].values[0]

stockprice_mc['StockIndexOpen'] = ibex_open
stockprice_mc['StockIndexClose'] = ibex_close

stockprice_mc.to_excel('stockprice_mc.xlsx', index=False)

stockprice_mc.head()
```
