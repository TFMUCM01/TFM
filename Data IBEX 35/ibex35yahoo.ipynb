```python
# Diccionario para guardar los ratios de cada ticker
ratios_dict = {}

for ti in tickers:
    ticker = yf.Ticker(ti)
    info = ticker.info

    # Lista de ratios comunes disponibles en yfinance
    ratios = {
        "PER (Price/Earnings)": info.get("trailingPE"),
        "Forward PER": info.get("forwardPE"),
        "PEG Ratio": info.get("pegRatio"),
        "Price/Sales": info.get("priceToSalesTrailing12Months"),
        "Price/Book": info.get("priceToBook"),
        "Return on Equity": info.get("returnOnEquity"),
        "Return on Assets": info.get("returnOnAssets"),
        "Debt/Equity": info.get("debtToEquity"),
        "Current Ratio": info.get("currentRatio"),
        "Quick Ratio": info.get("quickRatio"),
        "Gross Margins": info.get("grossMargins"),
        "Operating Margins": info.get("operatingMargins"),
        "Profit Margins": info.get("profitMargins"),
        "Beta": info.get("beta"),
        "Dividend Yield": info.get("dividendYield"),
    }

    # Convierte a DataFrame para visualizar mejor
    df_ratios = pd.DataFrame(list(ratios.items()), columns=["Ratio", "Valor"])
    ratios_dict[ti] = df_ratios

# Mostrar ratios de cada ticker
for ti, df_ratios in ratios_dict.items():
    print(f"Ratios para {ti}:")
    display(df_ratios)
```

```python
# Diccionario para guardar los DataFrames de cada tipo de informe por ticker
balances = {}
financieros = {}
flujos = {}

for ti in tickers:
    ticker = yf.Ticker(ti)
    # Balance general
    balance_sheet = ticker.balance_sheet
    balances[ti] = balance_sheet
    # Estado de resultados
    financials = ticker.financials
    financieros[ti] = financials
    # Flujos de caja
    cashflow = ticker.cashflow
    flujos[ti] = cashflow

# Mostrar los resultados
for ti in tickers:
    print(f"--- {ti} BALANCE SHEET ---")
    display(balances[ti])
    print(f"--- {ti} FINANCIALS ---")
    display(financieros[ti])
    print(f"--- {ti} CASHFLOW ---")
    display(flujos[ti])
```
