```python
import time
import requests
from bs4 import BeautifulSoup
from urllib import robotparser
from datetime import datetime

# 1. Configurar el parser de robots.txt
rp = robotparser.RobotFileParser()
rp.set_url("https://elpais.com/robots.txt")
rp.read()

user_agent = "MiScraperBot"
base_url = "https://elpais.com/economia/"
start_date = datetime(2025, 1, 1)
end_date = datetime(2025, 6, 12)

# 2. Verificar permiso
if not rp.can_fetch(user_agent, base_url):
    raise Exception("üõë La secci√≥n /economia/ est√° bloqueada por robots.txt")

# 3. Hacer peticiones con respeto
headers = {"User-Agent": user_agent}
response = requests.get(base_url, headers=headers, timeout=10)
soup = BeautifulSoup(response.text, "lxml")

# 4. Buscar enlaces y sus fechas
news = []
for article in soup.find_all("article"):
    link = article.find("a", href=True)
    time_tag = article.find("time", {"datetime": True})
    if link and time_tag:
        url = link["href"]
        if not url.startswith("http"):
            url = "https://elpais.com" + url
        try:
            pub = datetime.fromisoformat(time_tag["datetime"][:10])
            if start_date <= pub <= end_date:
                title = link.get_text(strip=True)
                news.append({"fecha": str(pub.date()), "titulo": title, "url": url})
        except Exception:
            continue

# 5. Mostrar resultados en consola
for n in news:
    print(f"{n['fecha']} - {n['titulo']}")
    print(f"‚Üí {n['url']}\n")

# 6. Evitar sobrecargar el servidor
time.sleep(2)
```

```python
from datetime import datetime, timedelta
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Rango de fechas para todo 2024
fecha_inicio = datetime.strptime("20240101", "%Y%m%d")
fecha_fin = datetime.strptime("20241231", "%Y%m%d")
original_url = 'https://elpais.com/economia/'
resultados = []

# Par√°metros para robustez
WAYBACK_TIMEOUT = 30      # segundos para la API de Wayback
SNAPSHOT_TIMEOUT = 30     # segundos para la descarga de la p√°gina archivada
RETRIES = 3               # reintentos por d√≠a si falla la conexi√≥n
SLEEP_BETWEEN_DAYS = 2    # segundos entre d√≠as para no saturar

fecha = fecha_inicio
while fecha <= fecha_fin:
    fecha_str = fecha.strftime("%Y%m%d")
    wayback_api = f'https://archive.org/wayback/available?url={original_url}&timestamp={fecha_str}'
    success = False
    for intento in range(RETRIES):
        try:
            res = requests.get(wayback_api, timeout=WAYBACK_TIMEOUT)
            data = res.json()
            if 'archived_snapshots' in data and data['archived_snapshots']:
                snapshot_url = data['archived_snapshots']['closest']['url']
                if snapshot_url.startswith("http://"):
                    snapshot_url = snapshot_url.replace("http://", "https://")
                try:
                    page = requests.get(snapshot_url, timeout=SNAPSHOT_TIMEOUT)
                    soup = BeautifulSoup(page.content, 'html.parser')
                    titulares = soup.find_all(['h2', 'h3'])
                    for t in titulares:
                        texto = t.get_text(strip=True)
                        if texto:
                            resultados.append({
                                "fecha": fecha_str,
                                "titular": texto,
                                "url_archivo": snapshot_url
                            })
                    success = True
                    break  # Sal del bucle de reintentos si fue exitoso
                except Exception as e:
                    print(f"Error accediendo a {snapshot_url} (intento {intento+1}): {e}")
            else:
                print(f"No hay snapshot para {fecha_str}")
                success = True  # No hay snapshot, pero no es error de conexi√≥n
                break
        except Exception as e:
            print(f"Error consultando Wayback para {fecha_str} (intento {intento+1}): {e}")
            time.sleep(3)  # Espera antes de reintentar
    if not success:
        print(f"Fallo persistente en {fecha_str}, se omite el d√≠a.")
    time.sleep(SLEEP_BETWEEN_DAYS)
    fecha += timedelta(days=1)

# DataFrame con los titulares
df_titulares = pd.DataFrame(resultados)
print(df_titulares.head())
print(f"Total de titulares recopilados: {len(df_titulares)}")

# Si quieres guardar el resultado:
df_titulares.to_csv("elpais_economia_2024.csv", index=False)
```
